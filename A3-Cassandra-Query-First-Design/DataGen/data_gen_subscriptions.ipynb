```python
##!pip install Faker;
import pandas as pd;
import numpy as np;
import random;
from faker import Faker;
from datetime import datetime, timezone
```

```python
users_table = pd.read_csv("../DB_Table/Users_Table.csv")
userids = users_table["user_id"]
```

```python
columns = ['subscription_id', 'subscriber_id', 'subscribed_to_id', 'subscription_date']
df_subscriptions = pd.DataFrame(columns=columns)
```

```python
# Iterate over each user ID , we are taking only 175 user IDs
for user_id in userids[:175]:
    # Generate a random number from 20 to 100 for each user ID
    random_number = random.randint(5, 50)
    for i in range(0, random_number):
        x = user_id
        y = random.choice(userids)
        if(x != y ):
            new_row = pd.Series(data = {'subscription_id':None, 'subscriber_id':x, 'subscribed_to_id':y, 'subscription_date':None}, name = None)
            #print(f"User ID {user_id}: subs {random.choice(userids)}")
            df_subscriptions = df_subscriptions.append(new_row, ignore_index=True)
```
